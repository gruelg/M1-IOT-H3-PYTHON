Gain de force,HP et speed en fonction du gain d'experience 

```python
import pandas as pd 
```

```python
DataPokemon = pd.read_csv('pokemon.csv')
```

```python
DataPokemon.shape
```

```python
DataPokemon.head(10)
```

```python
DataPokemon.isna()
```

```python
DataPokemon['type2'].fillna(value='None', inplace=True)
```

```python
DataPokemon.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
DataPokemon.dtypes
```

```python
x=DataPokemon[['attack','hp','speed']]
y=DataPokemon['experience_growth']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=42)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression 
clf = LinearRegression().fit(x_train, y_train)
```

```python
clf.predict(x_test)
```

```python
clf.score(x_test,y_test)
```

```python
clf.coef_
```
